```python
class BillApp:
    
    def __init__(self, root):
        
        self.root = root
        self.root.geometry('1350x700+0+0')
        self.root.title('Restaurant Billing')
        
        title = Label(self.root, text='Restaurant Billing', bd=12, relief=GROOVE, bg='#6B3074', fg='orange', font=('times new roman', 30, 'bold'), pady=2).pack(fill=X)
        
        #-----------------------------------------------------------
        
        self.c_name = StringVar()
        self.c_phone = StringVar()
        
        self.bill_no = StringVar()
        self.bill_no.set(str(random.randint(1000,9999)))
        
        self.search = StringVar()
        
        self.roll = IntVar()
        self.bread = IntVar()
        self.fries = IntVar()
        self.wings = IntVar()
        self.potato = IntVar()
        self.soup = IntVar()
        
        self.handi = IntVar()
        self.biryani = IntVar()
        self.karahi = IntVar()
        self.tikka = IntVar()
        self.malai = IntVar()
        self.haleem = IntVar()
        
        self.kitkat = IntVar()
        self.crunch = IntVar()
        self.mars = IntVar()
        self.toblerone = IntVar()
        self.dairymilk = IntVar()
        self.bounty = IntVar()
        
        self.starter_price = StringVar()
        self.maindish_price = StringVar()
        self.desert_price = StringVar()
        
        self.starter_tax = StringVar()
        self.maindish_tax = StringVar()
        self.desert_tax = StringVar()
        
        self.total_price = StringVar()
        
        #-----------------------------------------------------------
        
        frame1 = LabelFrame(self.root, text='Customer Details', bd=12, relief=GROOVE, bg='#6B3074', fg='orange', font=('times new roman', 15, 'bold'))
        frame1.place(x=0, y=80, relwidth=1)
        
        lbl1 = Label(frame1, text='Customer Name', bg='#6B3074', fg='white', font=('times new roman', 18, 'bold')).grid(row=0,column=0,padx=20,pady=5)
        lbl1_txt = Entry(frame1, textvariable=self.c_name , font=('times new roman', 15), bd=7, relief=SUNKEN).grid(row=0,column=1,padx=5,pady=10)
        
        lbl2 = Label(frame1, text='Phone No.', bg='#6B3074', fg='white', font=('times new roman', 18, 'bold')).grid(row=0,column=2,padx=20,pady=5)
        lbl2_txt = Entry(frame1, textvariable=self.c_phone , font=('times new roman', 15), bd=7, relief=SUNKEN).grid(row=0,column=3,padx=5,pady=10)
        
        lbl3 = Label(frame1, text='Bill No.', bg='#6B3074', fg='white', font=('times new roman', 18, 'bold')).grid(row=0,column=4,padx=20,pady=5)
        lbl3_txt = Entry(frame1, textvariable=self.search , font=('times new roman', 15), bd=7, relief=SUNKEN).grid(row=0,column=5,padx=5,pady=10)
        lbl3_btn = Button(frame1, text='Search', command=self.search_bill, width=8, bd=7, font=('times new roman', 12, 'bold')).grid(row=0,column=6,padx=20)
        
        #-----------------------------------------------------------
        
        frame2 = LabelFrame(self.root, text='Starters', labelanchor='n' , bd=12, relief=GROOVE, bg='#6B3074', fg='orange', font=('times new roman', 15, 'bold'))
        frame2.place(x=0, y=180, width=330, height=380)
        
        f2_itm1 = Label(frame2, text='Spring Roll', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=0,column=0,padx=10,pady=10,sticky='w')
        f2_itm1_txt = Entry(frame2, textvariable=self.roll , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=0,column=1,padx=10,pady=10)
        
        f2_itm2 = Label(frame2, text='Garlic Bread', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=1,column=0,padx=10,pady=10,sticky='w')
        f2_itm2_txt = Entry(frame2, textvariable=self.bread , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=1,column=1,padx=10,pady=10)

        f2_itm3 = Label(frame2, text='French Fries', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=2,column=0,padx=10,pady=10,sticky='w')
        f2_itm3_txt = Entry(frame2, textvariable=self.fries , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=2,column=1,padx=10,pady=10)

        f2_itm4 = Label(frame2, text='Chicken Wings', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=3,column=0,padx=10,pady=10,sticky='w')
        f2_itm4_txt = Entry(frame2, textvariable=self.wings , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=3,column=1,padx=10,pady=10)

        f2_itm5 = Label(frame2, text='Baked Potatoes', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=4,column=0,padx=10,pady=10,sticky='w')
        f2_itm5_txt = Entry(frame2, textvariable=self.potato , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=4,column=1,padx=10,pady=10)

        f2_itm6 = Label(frame2, text='Special Soup', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=5,column=0,padx=10,pady=10,sticky='w')
        f2_itm6_txt = Entry(frame2, textvariable=self.soup , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=5,column=1,padx=10,pady=10)
        
        #-----------------------------------------------------------
        
        frame3 = LabelFrame(self.root, text='Main Dishes', labelanchor='n' , bd=12, relief=GROOVE, bg='#6B3074', fg='orange', font=('times new roman', 15, 'bold'))
        frame3.place(x=340, y=180, width=330, height=380)
        
        f3_itm1 = Label(frame3, text='Chkn Handi 1/2', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=0,column=0,padx=10,pady=10,sticky='w')
        f3_itm1_txt = Entry(frame3, textvariable=self.handi , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=0,column=1,padx=10,pady=10)
        
        f3_itm2 = Label(frame3, text='Chkn Biryani 1/2', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=1,column=0,padx=10,pady=10,sticky='w')
        f3_itm2_txt = Entry(frame3, textvariable=self.biryani , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=1,column=1,padx=10,pady=10)

        f3_itm3 = Label(frame3, text='Chkn Karahi 1/2', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=2,column=0,padx=10,pady=10,sticky='w')
        f3_itm3_txt = Entry(frame3, textvariable=self.karahi , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=2,column=1,padx=10,pady=10)

        f3_itm4 = Label(frame3, text='Tikka', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=3,column=0,padx=10,pady=10,sticky='w')
        f3_itm4_txt = Entry(frame3, textvariable=self.tikka , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=3,column=1,padx=10,pady=10)

        f3_itm5 = Label(frame3, text='Malai Boti', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=4,column=0,padx=10,pady=10,sticky='w')
        f3_itm5_txt = Entry(frame3, textvariable=self.malai , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=4,column=1,padx=10,pady=10)

        f3_itm6 = Label(frame3, text='Haleem 1/2', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=5,column=0,padx=10,pady=10,sticky='w')
        f3_itm6_txt = Entry(frame3, textvariable=self.haleem , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=5,column=1,padx=10,pady=10)
        
        #-----------------------------------------------------------
        
        frame4 = LabelFrame(self.root, text='Desserts', labelanchor='n', bd=12, relief=GROOVE, bg='#6B3074', fg='orange', font=('times new roman', 15, 'bold'))
        frame4.place(x=680, y=180, width=330, height=380)
        
        f4_itm1 = Label(frame4, text='KitKat surprise', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=0,column=0,padx=10,pady=10,sticky='w')
        f4_itm1_txt = Entry(frame4, textvariable=self.kitkat , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=0,column=1,padx=10,pady=10)
        
        f4_itm2 = Label(frame4, text='Crunch icecream', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=1,column=0,padx=10,pady=10,sticky='w')
        f4_itm2_txt = Entry(frame4, textvariable=self.crunch , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=1,column=1,padx=10,pady=10)

        f4_itm3 = Label(frame4, text='Mars delight', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=2,column=0,padx=10,pady=10,sticky='w')
        f4_itm3_txt = Entry(frame4, textvariable=self.mars , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=2,column=1,padx=10,pady=10)

        f4_itm4 = Label(frame4, text='Toblerone', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=3,column=0,padx=10,pady=10,sticky='w')
        f4_itm4_txt = Entry(frame4, textvariable=self.toblerone , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=3,column=1,padx=10,pady=10)

        f4_itm5 = Label(frame4, text='DairyMilk Pie', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=4,column=0,padx=10,pady=10,sticky='w')
        f4_itm5_txt = Entry(frame4, textvariable=self.dairymilk , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=4,column=1,padx=10,pady=10)

        f4_itm6 = Label(frame4, text='Bounty delight', bg='#6B3074', fg='lightgreen',font=('times new roman', 16, 'bold')).grid(row=5,column=0,padx=10,pady=10,sticky='w')
        f4_itm6_txt = Entry(frame4, textvariable=self.bounty , font=('times new roman', 15), bd=5, relief=SUNKEN, width=10).grid(row=5,column=1,padx=10,pady=10)
        
        #-----------------------------------------------------------
        
        frame5 = Frame(self.root, bd=10, relief=GROOVE)
        frame5.place(x=1020, y=180, width=330, height=380)
        
        f5_title = Label(frame5, text='Bill', bd=7, relief=GROOVE, font=('arial', 15)).pack(fill=X)
        
        f5_scroll = Scrollbar(frame5, orient=VERTICAL)
        f5_scroll.pack(side=RIGHT, fill=Y)
        
        self.billarea = Text(frame5, yscrollcommand=f5_scroll.set)
        f5_scroll.config(command=self.billarea.yview)
        self.billarea.pack(fill=BOTH, expand=1)
        
         #-----------------------------------------------------------
            
        frame6 = LabelFrame(self.root, text='Bill Menu', labelanchor='n', bd=12, relief=GROOVE, bg='#6B3074', fg='orange', font=('times new roman', 15, 'bold'))
        frame6.place(x=0, y=570, relwidth=1, height=140)  
        
        f6_itm1 = Label(frame6, text='Total Starter Price', bg='#6B3074', fg='white',font=('times new roman', 14, 'bold')).grid(row=0,column=0,padx=20,pady=2,sticky='w')
        f6_itm1_txt = Entry(frame6, textvariable=self.starter_price , font=('times new roman', 14), bd=5, relief=SUNKEN, width=14).grid(row=0,column=1,padx=20,pady=1)
        
        f6_itm2 = Label(frame6, text='Total Main Dish Price', bg='#6B3074', fg='white',font=('times new roman', 14, 'bold')).grid(row=1,column=0,padx=20,pady=2,sticky='w')
        f6_itm2_txt = Entry(frame6, textvariable=self.maindish_price , font=('times new roman', 14), bd=5, relief=SUNKEN, width=14).grid(row=1,column=1,padx=20,pady=1)

        f6_itm3 = Label(frame6, text='Total Dessert Price', bg='#6B3074', fg='white',font=('times new roman', 14, 'bold')).grid(row=2,column=0,padx=20,pady=2,sticky='w')
        f6_itm3_txt = Entry(frame6, textvariable=self.desert_price , font=('times new roman', 14), bd=5, relief=SUNKEN, width=14).grid(row=2,column=1,padx=20,pady=1) 
        
        f6_itm4 = Label(frame6, text='Starter Tax', bg='#6B3074', fg='white',font=('times new roman', 14, 'bold')).grid(row=0,column=2,padx=20,pady=2,sticky='w')
        f6_itm4_txt = Entry(frame6, textvariable=self.starter_tax , font=('times new roman', 14), bd=5, relief=SUNKEN, width=14).grid(row=0,column=3,padx=20,pady=1)
        
        f6_itm5 = Label(frame6, text='Main Dish Tax', bg='#6B3074', fg='white',font=('times new roman', 14, 'bold')).grid(row=1,column=2,padx=20,pady=2,sticky='w')
        f6_itm5_txt = Entry(frame6, textvariable=self.maindish_tax , font=('times new roman', 14), bd=5, relief=SUNKEN, width=14).grid(row=1,column=3,padx=20,pady=1)

        f6_itm6 = Label(frame6, text='Dessert Tax', bg='#6B3074', fg='white',font=('times new roman', 14, 'bold')).grid(row=2,column=2,padx=20,pady=2,sticky='w')
        f6_itm6_txt = Entry(frame6, textvariable=self.desert_tax , font=('times new roman', 14), bd=5, relief=SUNKEN, width=14).grid(row=2,column=3,padx=20,pady=1)
 
        f6_itm7 = Frame(frame6, bd=7, relief=GROOVE)
        f6_itm7.place(x=750, width=575, height=104)
        
        total_btn = Button(f6_itm7, text='Total', command=self.total, bg='purple', fg='white', bd=5, width=10, height=2, font='arial 14').grid(row=0, column=0, padx=4, pady=9)
        
        bill_btn = Button(f6_itm7, text='Generate Bill', command=self.bill, bg='purple', fg='white', bd=5, width=11, height=2, font='arial 14').grid(row=0, column=1, padx=4, pady=9)

        clear_btn = Button(f6_itm7, text='Clear', command=self.clear_data, bg='purple', fg='white', bd=5, width=10, height=2, font='arial 14').grid(row=0, column=2, padx=4, pady=9)
        
        exit_btn = Button(f6_itm7, text='Exit', command=self.exit, bg='purple', fg='white', bd=5, width=10, height=2, font='arial 14').grid(row=0, column=3, padx=4, pady=9)
        
        #----------------------------------------------------------------
        
    def total(self):
        
        s_p = (self.roll.get()*80)+(self.bread.get()*50)+(self.fries.get()*40)+(self.wings.get()*50)+(self.potato.get()*100)+(self.soup.get()*100)
        self.starter_price.set(str(s_p))
        s_p_tax = round(s_p*0.05,2)
        self.starter_tax.set(str(s_p_tax))
        
        m_p = (self.handi.get()*250)+(self.biryani.get()*250)+(self.tikka.get()*160)+(self.malai.get()*180)+(self.haleem.get()*170)+(self.karahi.get()*260)
        self.maindish_price.set(str(m_p))
        m_p_tax = round(m_p*0.1,2)
        self.maindish_tax.set(str(m_p_tax))
        
        d_p = (self.kitkat.get()*150)+(self.crunch.get()*50)+(self.mars.get()*100)+(self.toblerone.get()*200)+(self.dairymilk.get()*120)+(self.bounty.get()*120)
        self.desert_price.set(str(d_p))
        d_p_tax = round(d_p*0.05,2)
        self.desert_tax.set(str(d_p_tax))
        
        self.total_price.set(s_p+s_p_tax+m_p+m_p_tax+d_p+d_p_tax)
        
        
    def bill(self):
        
        if self.c_name.get() == '' or self.c_phone.get() == '':
            messagebox.showerror('Error','Incomplete Customer Details.\nKindly enter customer details')
            
        elif self.starter_price.get() < '0.01' and self.maindish_price.get() < '0.01' and self.desert_price.get() < '0.1':
            messagebox.showerror('Error','No product purchased')
            
        else:
            self.billarea.delete('1.0',END)
            self.billarea.insert(END,'\tWelcome to our Restaurant')
            self.billarea.insert(END,'\n\t==========================')
            self.billarea.insert(END,f'\n Bill No : {self.bill_no.get()}')
            self.billarea.insert(END,f'\n Customer Name : {self.c_name.get()}')
            self.billarea.insert(END,f'\n Phone : {self.c_phone.get()}')
            self.billarea.insert(END,'\n====================================')
            self.billarea.insert(END,'\n Product\t\tQTY\t    Price')
            self.billarea.insert(END,'\n====================================')

         #---------------------------------------------------------------------------------

            if self.roll.get()>0:
                self.billarea.insert(END,f'\n Spring Roll    \t\t{self.roll.get()} \t     {self.roll.get()*80}')

            if self.bread.get()>0:
                self.billarea.insert(END,f'\n Garlic Bread   \t\t{self.bread.get()} \t     {self.bread.get()*50}')

            if self.fries.get()>0:
                self.billarea.insert(END,f'\n French Fries   \t\t{self.fries.get()} \t     {self.fries.get()*40}')

            if self.wings.get()>0:
                self.billarea.insert(END,f'\n Chicken Wings  \t\t{self.wings.get()} \t     {self.wings.get()*50}')

            if self.potato.get()>0:
                self.billarea.insert(END,f'\n Baked Potatoes \t\t{self.potato.get()} \t     {self.potato.get()*100}')

            if self.soup.get()>0:
                self.billarea.insert(END,f'\n Special Soup   \t\t{self.soup.get()} \t     {self.soup.get()*100}')

         #---------------------------------------------------------------------------------

            if self.handi.get()>0:
                self.billarea.insert(END,f'\n Chicken handi  \t\t{self.handi.get()} \t     {self.handi.get()*250}')

            if self.biryani.get()>0:
                self.billarea.insert(END,f'\n Chicken biryani\t\t{self.biryani.get()} \t     {self.biryani.get()*250}')

            if self.karahi.get()>0:
                self.billarea.insert(END,f'\n Chiken karahi  \t\t{self.karahi.get()} \t     {self.karahi.get()*260}')

            if self.tikka.get()>0:
                self.billarea.insert(END,f'\n Chicken tikka  \t\t{self.tikka.get()} \t     {self.tikka.get()*160}')

            if self.malai.get()>0:
                self.billarea.insert(END,f'\n Malai boti     \t\t{self.malai.get()} \t     {self.malai.get()*180}')

            if self.haleem.get()>0:
                self.billarea.insert(END,f'\n Haleem         \t\t{self.haleem.get()} \t     {self.haleem.get()*170}')

         #---------------------------------------------------------------------------------

            if self.kitkat.get()>0:
                self.billarea.insert(END,f'\n KitKat Surprise\t\t{self.kitkat.get()} \t     {self.kitkat.get()*150}')

            if self.crunch.get()>0:
                self.billarea.insert(END,f'\n Crunch Kulfi   \t\t{self.crunch.get()} \t     {self.crunch.get()*50}')

            if self.mars.get()>0:
                self.billarea.insert(END,f'\n Mars Delight   \t\t{self.mars.get()} \t     {self.mars.get()*100}')

            if self.toblerone.get()>0:
                self.billarea.insert(END,f'\n Toblerone      \t\t{self.toblerone.get()} \t     {self.toblerone.get()*200}')

            if self.dairymilk.get()>0:
                self.billarea.insert(END,f'\n Dairy Milk Pie \t\t{self.dairymilk.get()} \t     {self.dairymilk.get()*120}')

            if self.bounty.get()>0:
                self.billarea.insert(END,f'\n Bounty Delight \t\t{self.bounty.get()} \t     {self.bounty.get()*120}')

            #---------------------------------------------------------------------------------

            self.billarea.insert(END,'\n====================================')

            if self.starter_tax.get() > '0.0':
                self.billarea.insert(END,f'\n Starter Tax\t\t\t     {self.starter_tax.get()}')

            if self.maindish_tax.get() > '0.0':
                self.billarea.insert(END,f'\n Main Dish Tax\t\t\t     {self.maindish_tax.get()}')

            if self.desert_tax.get() > '0.0':
                self.billarea.insert(END,f'\n Dessert Tax\t\t\t     {self.desert_tax.get()}')

            #---------------------------------------------------------------------------------

            self.billarea.insert(END,'\n====================================')
            self.billarea.insert(END,f'\n Total\t\t\t Rs. {self.total_price.get()}')
            self.billarea.insert(END,'\n====================================')
            
            self.get_bill()
            
    def search_bill(self):
        exist = False
        for i in os.listdir('RestaurantBills/'):
            if i.split('.')[0] == self.search.get():
                exist = True
                file = open(f'RestaurantBills/{i}','r')
                self.billarea.delete('1.0',END)
                for line in file:
                    self.billarea.insert(END,line)
                file.close()
                
        if not exist:
            messagebox.showerror('Invalid Bill No','Bill not found')
    
    def get_bill(self):     
        
        msg_per = messagebox.askyesno('Save Bill','Do you want to save bill?')
        
        if msg_per:
            bill_data = self.billarea.get('1.0',END)

            file = open('RestaurantBills/'+str(self.bill_no.get())+'.txt','w')
            file.write(str(bill_data))
            file.close()
            
            messagebox.showinfo('Bill Saved',f'Bill No. {self.bill_no.get()} has been saved')
        
        else:
            return
        
    def clear_data(self):

        choice = messagebox.askyesno('Exit','Do you want to clear data?')
        if choice:
            self.c_name.set('')
            self.c_phone.set('')
            self.bill_no.set('')
            self.search.set('')

            self.roll.set(0)
            self.bread.set(0)
            self.fries.set(0)
            self.wings.set(0)
            self.potato.set(0)
            self.soup.set(0)

            self.handi.set(0)
            self.biryani.set(0)
            self.karahi.set(0)
            self.tikka.set(0)
            self.malai.set(0)
            self.haleem.set(0)

            self.kitkat.set(0)
            self.crunch.set(0)
            self.mars.set(0)
            self.toblerone.set(0)
            self.dairymilk.set(0)
            self.bounty.set(0)

            self.starter_price.set('')
            self.maindish_price.set('')
            self.desert_price.set('')

            self.starter_tax.set('')
            self.maindish_tax.set('')
            self.desert_tax.set('')
            self.total_price.set('')
        
        else:
            return
        
    def exit(self):
        choice = messagebox.askyesno('Exit','Do you want to exit?')
        if choice:
            self.root.destroy()
        else:
            return
```
